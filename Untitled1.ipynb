```python
sns.set_style('white')
sns.set_context('notebook')
sns.set_palette('dark')
f = plt.figure(figsize=(16,12))
sns.boxplot(x='Revenue_type', y='Revenue', 
            hue='New_Zone', data=plot_data);
```

```python
plt.figure(figsize=(18,10))
n_df.boxplot();
plt.xticks(rotation=60, size=10)
```

```python
skew_limit = 0.75
skew_vals = n_df.skew()
```

```python
skew_cols = (abs(skew_vals)
             .sort_values(ascending=False)
             .to_frame()
             .rename(columns={0:'Skew'})
             .query('abs(Skew) > {}'.format(skew_limit)))

skew_cols
```

```python
for col in skew_cols.index.values:
    n_df[col] = n_df[col].apply(np.log1p) # for coe_ np.expm1(Y)
```

```python
n_df.isnull().sum().sort_values()
```

```python
n_df = n_df.fillna(0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, MinMaxScaler
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
n_df.columns
```

```python
LR = LinearRegression()
scaler = MinMaxScaler()
X = n_df.drop(['Total Revenues','Room Revenue', 'Food & Beverage Revenue', 'Other Revenue'], axis=1)
y = n_df['Total Revenues']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, random_state=42)
```

```python
s_X_train = scaler.fit_transform(X_train)
s_X_test = scaler.transform(X_test)
```

```python
LR.fit(s_X_train, y_train)
pred = LR.predict(s_X_test)
```

```python
pd.set_option('float_format', '{:f}'.format)
```

```python
mean_squared_error(y_test, pred)
```

```python
r2_score(y_test, pred)
```

```python
pd.DataFrame(list(zip(X.columns, np.expm1(LR.coef_))))
```

```python
kf = KFold(shuffle=True, random_state=72018, n_splits=3)
```

```python
for train_index, test_index in kf.split(X):
    print("Train index:", train_index[:10], len(train_index))
    print("Test index:",test_index[:10], len(test_index))
    print('')
```

```python
alphas = np.geomspace(1e-9, 1e0, num=10)
```

```python
s = StandardScaler()
```

```python
scores = []
coefs = []
for alpha in alphas:
    las = Lasso(alpha=alpha, max_iter=100000)
    
    estimator = Pipeline([
        ("scaler", s),
        ("lasso_regression", las)])

    predictions = cross_val_predict(estimator, X, y, cv = kf)
    
    score = r2_score(y, predictions)
    
    scores.append(score)
```

```python
list(zip(alphas,scores))
```

```python
Lasso(alpha=1e-6).fit(X, y).coef_
```

```python
pf = PolynomialFeatures(degree=3)

scores = []
alphas = np.geomspace(0.06, 6.0, 20)
for alpha in alphas:
    las = Lasso(alpha=alpha, max_iter=100000)
    
    estimator = Pipeline([
        ("scaler", s),
        ("make_higher_degree", pf),
        ("lasso_regression", las)])

    predictions = cross_val_predict(estimator, X, y, cv = kf)
    
    score = r2_score(y, predictions)
    
    scores.append(score)
    
```

```python
list(zip(alphas,scores))
```

```python
plt.semilogx(alphas, scores);
```

```python
estimator = Pipeline([("scaler", StandardScaler()),
        ("polynomial_features", PolynomialFeatures()),
        ("lasso_regression", Lasso(alpha=alpha, max_iter=100000))])

params = {
    'polynomial_features__degree': [1, 2, 3],
    'lasso_regression__alpha': np.geomspace(0.06, 6.0, 20)
}

grid = GridSearchCV(estimator, params, cv=kf)
```

```python
grid.fit(X, y)
```

```python
grid.best_score_, grid.best_params_
```

```python
best_estimator = Pipeline([
                    ("scaler", s),
                    ("make_higher_degree", PolynomialFeatures(degree=1)),
                    ("lasso_regression", Lasso(alpha=0.06))])

best_estimator.fit(X, y)
best_estimator.score(X, y)
```

```python
best_estimator.named_steps["lasso_regression"].coef_
```

```python
estimator = Pipeline([("scaler", StandardScaler()),
        ("polynomial_features", PolynomialFeatures()),
        ("ridge_regression", Ridge())])

params = {
    'polynomial_features__degree': [1, 2, 3],
    'ridge_regression__alpha': np.geomspace(4, 20, 30)
}

grid = GridSearchCV(estimator, params, cv=kf)
```

```python
grid.fit(X, y)
```

```python
grid.best_score_, grid.best_params_ 
```

```python
y_predict = grid.predict(X)
```

```python
r2_score(y, y_predict)
```

```python
grid.best_estimator_.named_steps['ridge_regression'].coef_
```
